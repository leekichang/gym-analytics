```python
import cv2
import torch
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt

import config as cfg
```

```python
model = YOLO('yolo11x-pose.pt')
video = './data/single-arm-lat-pull-down-1.MOV'
results = model(source=video, save=True)
```

```python
confs = {k:[] for k in range(17)}
for frame_idx, r in enumerate(results):
    for k, c in enumerate(r.keypoints.conf[0]):
        confs[k].append(c.cpu())
```

```python
cmap = plt.get_cmap('tab20')
plt.figure(figsize=(16,6))
for idx, k in enumerate(confs):
    plt.plot(confs[k], label=cfg.label2key[idx], linewidth=2, color=cmap(idx))
plt.legend(ncol=9, loc='upper center', bbox_to_anchor=(0.5, 1.1))
plt.show()
```
